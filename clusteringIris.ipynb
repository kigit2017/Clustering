```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X,y = iris.data,iris.target
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

k value = 4

```python
kmeans= KMeans(n_clusters=4,random_state=42)
kmeans.fit(X)
```

```python
cluster_labels = kmeans.labels_
cluster_center = kmeans.cluster_centers_
```

Scatter plot for Iris data  with k values

```python
plt.scatter(X[:,0],X[:,1],c=cluster_labels,cmap='viridis')
plt.scatter(cluster_center[:,0],cluster_center[:,1],c='red',marker='X',s=200,label='Centroid')
plt.legend()
plt.show()
```

Elbow curve- Plotting sumsofsquares of centroids and  cluster numbers.
K= range of clusters
For the range fit the data.
Get sum of scuares  of centroid distances.

```python
k_val = range(1,10)
wcss=[]
```

```python
for k in k_val:
  kmeans = KMeans(n_clusters=k,random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
wcss  # length of
```

plot wcss and K

K optimal value is 4 - the increase in k does not signicantly change the sum of squares ink, for k value of 4

```python
plt.figure(figsize=(8,6))

plt.plot(k_val,wcss,marker='o',color='b',linestyle='-')
plt.title('Elbow curve-Kmeans optimal value')
plt.ylabel('sum of squares in K')
plt.xticks(k_val)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
```

Draw dendrogram for iris data - hierarchical clusters(grouping and merging in a tree) - bottom up dendro

```python
plt.figure(figsize=(10,6))
linkage_matrix= linkage(X,method='ward')# ward is linkage criteria(sensible method, other methods,complete,mean)
dendrogram(linkage_matrix,p=5, truncate_mode='level',orientation='bottom',labels=None, distance_sort='descending')
# p =5 mas leaf,
plt.xlabel('Data points')
plt.ylabel('Distance')
```
